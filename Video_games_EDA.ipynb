```python
def summary(data):
    summ = pd.DataFrame()
    summ['dataTypes'] = data.dtypes
    summ['Total'] = data.count()
    summ['unique'] = data.nunique()
    summ['missing'] = data.isnull().sum()
    summ['missing%'] = data.isnull().sum() / len(data) * 100
    stats = data.describe().transpose()
    summ['min'] = stats['min']
    summ['max'] = stats['max']
    summ['avg'] = stats['mean']
    summ['standard deveiation'] = stats['std']
    return summ

summary(df).style.background_gradient(cmap = 'Greens').highlight_null(color='purple')
```

```python
categorical_summary = df.describe(include=['object'])

skewness = df[['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].skew()
kurtosis = df[['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].kurt()


platform_counts = df['Platform'].value_counts()
genre_counts = df['Genre'].value_counts()
publisher_counts = df['Publisher'].value_counts()
```

```python

print("\nCategorical Summary:\n", categorical_summary)
print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)
print("\nPlatform Counts:\n", platform_counts)
print("\nGenre Counts:\n", genre_counts)
print("\nPublisher Counts:\n", publisher_counts)
```

```python
# Visualize Kurtosis
plt.figure(figsize=(10, 6))
sns.barplot(x=kurtosis.index, y=kurtosis.values, palette="magma")
plt.axhline(3, color='red', linestyle='--')  # Kurtosis of normal distribution is 3
plt.title('Kurtosis of Columns')
plt.ylabel('Kurtosis')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=skewness.index, y=skewness.values, palette="viridis")
plt.axhline(0, color='red', linestyle='--')
plt.title('Skewness of Columns')
plt.ylabel('Skewness')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a DataFrame for combined skewness and kurtosis
import pandas as pd

stats_df = pd.DataFrame({'Skewness': skewness, 'Kurtosis': kurtosis})

# Plot side-by-side bar plots
stats_df.plot(kind='bar', subplots=True, layout=(1, 2), figsize=(14, 6), legend=True, color=['purple', 'orange'])
plt.suptitle('Skewness and Kurtosis of Columns')
plt.show()
```

```python
# Combined bar plot for skewness and kurtosis
plt.figure(figsize=(10, 6))
width = 0.4  # Width of the bars

# Positions of bars
r1 = range(len(skewness))
r2 = [x + width for x in r1]

# Plot both skewness and kurtosis side by side
plt.bar(r1, skewness.values, width=width, label='Skewness', color='purple')
plt.bar(r2, kurtosis.values, width=width, label='Kurtosis', color='orange')

# Add labels and titles
plt.title('Skewness and Kurtosis Comparison')
plt.xticks([r + width/2 for r in range(len(skewness))], skewness.index, rotation=45)
plt.axhline(0, color='red', linestyle='--', label='Zero Skewness')
plt.axhline(3, color='green', linestyle='--', label='Normal Kurtosis (3)')
plt.ylabel('Value')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group by 'Year' and sum sales columns
games_per_year = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum().reset_index()

print(games_per_year.head())
sales_per_year.plot(kind='line', figsize=(12,8), title='Total Sales Per Year by Region')
```

```python
platform_sales_per_year = df.groupby(['Year', 'Platform']).sum()['Global_Sales'].unstack().fillna(0)
platform_sales_per_year.plot(kind='line', stacked=False, figsize=(12,8), title='Platform Popularity Over Time')
```

```python
publisher_sales_per_year = df.groupby(['Year', 'Publisher']).sum()['Global_Sales'].unstack().fillna(0)
top_publishers = publisher_sales_per_year.sum().sort_values(ascending=False).head(10).index
publisher_sales_per_year[top_publishers].plot(kind='line', figsize=(12,8), title='Top Publishers Over Time')
```

```python
total_sales_by_region = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
total_sales_by_region.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8), title='Total Sales Distribution by Region')
```
